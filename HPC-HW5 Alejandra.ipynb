$\textbf{Point 1:}$ Plot the function $$f(x)=a\sin(\cos(v-wx))+bx-cx^2$$ with the parameter  s given in the file.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun(a,v,w,b,c,vec):
    return a*np.sin(np.cos(v-w*vec))+b*vec-c*(vec**2)
    
```

```python
vec=np.linspace(0.,20.,100)
solution=fun(0.3, 2/3.0, 2.0, 1/1.3, 1/30.0,vec)
plt.plot(vec,solution)
plt.title("Plot function $f(x)=a\sin(\cos(v-wx))+bx-cx^2$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.show()
```
